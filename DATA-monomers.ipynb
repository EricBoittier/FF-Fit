```python
from ff_energy.geometry import angle, dist


def get_internals_water(key, res):
    mask = structure_key_pairs[key].res_mask[res]
    water_mol = structure_key_pairs[key].xyzs[mask]
    water_mol
    b1 = water_mol[0] - water_mol[1]
    b2 = water_mol[0] - water_mol[2]
    a = angle(b1, b2)
    r1 = dist(b1)
    r2 = dist(b2)
    return a, r1, r2
```

```python
from scipy.optimize import minimize
from ff_energy.bonded_terms import harmonic, harmonic_angle

H2KCALMOL = 627.503


def energy_interal(kb, ka, r0, a0, a, r1, r2):
    E = 0
    E += harmonic(r1, kb, r0)
    E += harmonic(r2, kb, r0)
    E += harmonic_angle(a, ka, a0)
    return E


def get_loss_df(x):
    df = data_.monomers_df.copy()
    E_pred = []
    for a, r1, r2 in zip(df["a"], df["r1"], df["r2"]):
        a = np.deg2rad(a)
        E_pred.append(energy_interal(*x, a, r1, r2))
    df["E_pred"] = E_pred
    df["m_ENERGY"] = df["m_ENERGY"] * H2KCALMOL
    df["m_ENERGY"] = df["m_ENERGY"] - df["m_ENERGY"].min()
    df["SE"] = (df["E_pred"] - df["m_ENERGY"]) ** 2
    return df


def get_loss(x):
    df = get_loss_df(x)
    MSE = df["SE"].mean()
    return MSE


x0 = [1, 1, 1, 2]
res = minimize(get_loss, x0, tol=1e-6)
res
```
